```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import scipy.optimize as opt
from pandas_datareader import data as pdr
import fix_yahoo_finance as yf
import pytz
import pandas as pd
import sqlite3 as db
from urllib.request import urlopen
import matplotlib.pyplot as plt

df_spy = pdr.get_data_yahoo('SPY',start='2013-04-03',end='2018-04-03')
df_spy_price = df_spy['Adj Close'].pct_change().dropna(axis=0,how='any')
print(df_spy_price)
```

```python
sectors = ['XLF','VNQ','XLK','VGT','XLE','XLV','XLI','XLY','IBB','AMLP','XLP','GDX','VFH','XLU','VHT']
sec_etf = pdr.get_data_yahoo(sectors,start='2013-04-03', end='2018-04-03')
sectors_r = sec_etf['Adj Close'].pct_change().dropna(axis=0, how='any')
print(sectors_r)
```

```python
x, rnorm = opt.nnls((sectors_r+1).values, (df_spy_price+1).values)
#for i in range(15):
    #print(sectors_r[i],': ',round(x[i],4))
x
```

```python
etf_portfolio_r = sectors_r @ x
np.corrcoef(etf_portfolio_r,df_spy_price)
```

```python
fig, ax = plt.subplots(figsize = (20,10))
plt.plot(df_spy_price.index,(sectors_r@x),label = "ETF's return")
plt.plot(df_spy_price.index,df_spy_price,label = 'SPY return')
plt.xlabel('Time')
plt.ylabel('Daily Return')
plt.legend()
```
