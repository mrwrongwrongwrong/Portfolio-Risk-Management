```python
import requests
import pytz
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import math


hdr = {'User-Agent': 'Mozilla/5.0'}
req = requests.get("http://en.wikipedia.org/wiki/List_of_S%26P_500_companies", headers=hdr)
soup = BeautifulSoup(req.content, "html.parser")

table = soup.find('table', {'class': 'wikitable sortable'})
sector_T = dict()
for row in table.findAll('tr'):
    col = row.findAll('td')
    if len(col) > 0:
        sector = str(col[3].string.strip()).lower().replace(' ', '_')
        ticker = str(col[0].string.strip())
        if sector not in sector_T:
            sector_T[sector] = list()
        sector_T[sector].append(ticker)
        
for row in table.findAll('tr'):
    col = row.findAll('td')
    if len(col) > 0:
        print(col)
        
D = sector_T 
tickers = []
for i in D.values():
    tickers.extend(i)
tickers.sort()
for i in tickers:
    if '.' in i:
        tickers.remove(i)
```
