```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import pandas_datareader as pdr
from datetime import datetime

wmt = pdr.get_data_yahoo(symbols='WMT', start=datetime(2007, 1, 1), end=datetime(2017, 5, 10))
#wmt.rename_axis(None)
#print(wmt.dtypes)


df = pd.read_excel('Bulk Dividend Downloader.xlsm', sheetname='WMT',skiprows=1)

Date = df['Date']
print(wmt.ix[Date])
#print(df)
#print(Date[1])
#print(wmt.ix[Date[1]])
#temp = Date[1].date()
#print(df.dtypes)
f = 1
for i in range(len(Date)-1,-1,-1):
    #day1 = Date[i].date()
    #day1 = Date[i-1].date()
    d1 = df.iloc[i]['Dividends']
    p1 = wmt.iloc[i]['Close']
    p2 = wmt.iloc[i-1]['Close']
    f  = f*(1-(d1/p1))
    adj = p2*f
    print(adj)
    

#print(df.iloc[1])


```
